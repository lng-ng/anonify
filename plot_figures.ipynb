```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Accuracy

```python
data_path = "figure_data/classifiers/Accuracy.csv"
df = pd.read_csv(data_path)

subjects = ["NB", "DT", "KNN", "SVM", "RF", "LR", "AB", "BG"]
original = df['Original Dataset']
anonymized = df['k250-anonymized']
sampling90 = df['k250, Deletion10%']
sampling70 = df['k250, Deletion30%']
sampling50 = df['k250, Deletion50%']
sampling30 = df['k250, Deletion70%']

# Set the width of the bars
bar_width = 0.1

# Create an array of subject positions for the x-axis
subject_positions = np.arange(len(subjects))

# Create the grouped bar chart
bar_positions = [subject_positions + i * bar_width for i in range(6)]

# Set the new x-axis labels
group_labels = ["Original", "Anonymized", "Sampling90", "Sampling70", "Sampling50", "Sampling30"]

plt.figure(figsize=(8, 6))
# Create the grouped bars for each dataset
for i, data in enumerate([original, anonymized, sampling90, sampling70, sampling50, sampling30]):
    plt.bar(bar_positions[i], data, bar_width, label=group_labels[i])

# Set x-axis labels
plt.xticks(subject_positions + 2.5 * bar_width, subjects, fontsize=12)

# Set the chart labels and title
plt.xlabel('Classifiers', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
#plt.title('Accuracy Comparison by Dataset and Classifier')

# Add a legend
plt.legend(loc='lower right', fontsize=12)

# Display the chart
plt.show()
```

# NCP

```python
data_path = "figure_data/ncp_table.csv"
df = pd.read_csv(data_path)
k_values = [50,100,150,200,250,300,350,400,450,500]
data = [
    ("ARX", k_values, df.loc[df['Algorithm'] == 'arx'].sort_values(by=['k'], axis='rows')['ncp']),
    ('Mondrian', k_values, df.loc[df['Algorithm'] == 'mondrian'].sort_values(by=['k'], axis='rows')['ncp']),
    ('OKA', k_values, df.loc[df['Algorithm'] == 'oka'].sort_values(by=['k'], axis='rows')['ncp'])
]
# Create a line chart
plt.figure(figsize=(10, 6))

for algo, k_values, ncp_values in data:
    plt.plot(k_values, ncp_values, label=algo)

plt.xlabel("k values", fontsize=12)
plt.ylabel("NCP%", fontsize=12)
#plt.title("Comparison of Algorithms")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.ylim(0, 80)
plt.show()
```

# Age-Diabetes

```python
# Age ranges
age_ranges = ['[0, 20)', '[20, 40)', '[40, 60)', '[60, 80)', '[80, 100)']

data_path = "figure_data/age_diabetes_k250_del50.csv"
df = pd.read_csv(data_path, sep=";")
df0 = df.loc[df['diabetes'] == 0]
df1 = df.loc[df['diabetes'] == 1]

# Data for each age range (diabetes=1)
diabetes_1_original = df1['Original Dataset']
diabetes_1_k_anonymity = df1['Only K-Anonymity']
diabetes_1_random_deletion_average = df1['Random Deletion (average)']

# Data for each age range (diabetes=0)
diabetes_0_original = df0['Original Dataset']
diabetes_0_k_anonymity = df0['Only K-Anonymity']
diabetes_0_random_deletion_average = df0['Random Deletion (average)']

# Set the width of the bars
bar_width = 0.2

# Create an array of age range positions for the x-axis
age_range_positions = np.arange(len(age_ranges))

# Create the grouped bar chart for diabetes=1
plt.figure(figsize=(12, 6))
diabetes_1_original_bars = plt.bar(age_range_positions - bar_width, diabetes_1_original, bar_width, label='Original')
diabetes_1_k_anonymity_bars = plt.bar(age_range_positions, diabetes_1_k_anonymity, bar_width, label='Anonymized')
diabetes_1_random_deletion_average_bars = plt.bar(age_range_positions + bar_width, diabetes_1_random_deletion_average, bar_width, label='Sampling50')

# Calculate the centers of the bars for diabetes=1
diabetes_1_original_centers = [rect.get_x() + rect.get_width() / 2 for rect in diabetes_1_original_bars]
diabetes_1_k_anonymity_centers = [rect.get_x() + rect.get_width() / 2 for rect in diabetes_1_k_anonymity_bars]
diabetes_1_random_deletion_average_centers = [rect.get_x() + rect.get_width() / 2 for rect in diabetes_1_random_deletion_average_bars]

# Draw lines connecting the centers of bars for diabetes=1
plt.plot(diabetes_1_original_centers, diabetes_1_original, marker='o', color='blue')
plt.plot(diabetes_1_k_anonymity_centers, diabetes_1_k_anonymity, marker='o', color='orange')
plt.plot(diabetes_1_random_deletion_average_centers, diabetes_1_random_deletion_average, marker='o', color='green')

# Set x-axis labels
plt.xticks(age_range_positions, age_ranges, fontsize=12)

# Set the chart labels and title for diabetes=1
plt.xlabel('Age Ranges', fontsize=12)
plt.ylabel('Number of people with diabetes', fontsize=12)
#plt.title('Grouped Bar Chart (Diabetes=1) by Age Ranges')

# Add a legend for diabetes=1
plt.legend(loc='upper left', fontsize=12)

# Show the plot for diabetes=1
plt.show()

# Create the grouped bar chart for diabetes=0
plt.figure(figsize=(12, 6))
diabetes_0_original_bars = plt.bar(age_range_positions - bar_width, diabetes_0_original, bar_width, label='Original')
diabetes_0_k_anonymity_bars = plt.bar(age_range_positions, diabetes_0_k_anonymity, bar_width, label='Anonymized')
diabetes_0_random_deletion_average_bars = plt.bar(age_range_positions + bar_width, diabetes_0_random_deletion_average, bar_width, label='Sampling50')

# Calculate the centers of the bars for diabetes=0
diabetes_0_original_centers = [rect.get_x() + rect.get_width() / 2 for rect in diabetes_0_original_bars]
diabetes_0_k_anonymity_centers = [rect.get_x() + rect.get_width() / 2 for rect in diabetes_0_k_anonymity_bars]
diabetes_0_random_deletion_average_centers = [rect.get_x() + rect.get_width() / 2 for rect in diabetes_0_random_deletion_average_bars]

# Draw lines connecting the centers of bars for diabetes=0
plt.plot(diabetes_0_original_centers, diabetes_0_original, marker='o', color='blue')
plt.plot(diabetes_0_k_anonymity_centers, diabetes_0_k_anonymity, marker='o', color='orange')
plt.plot(diabetes_0_random_deletion_average_centers, diabetes_0_random_deletion_average, marker='o', color='green')

# Set x-axis labels
plt.xticks(age_range_positions, age_ranges, fontsize=12)

# Set the chart labels and title for diabetes=0
plt.xlabel('Age Ranges', fontsize=12)
plt.ylabel('Number of people with no diabetes', fontsize=12)
#plt.title('Grouped Bar Chart (Diabetes=0) by Age Ranges')

# Add a legend for diabetes=0
plt.legend(loc='upper right', fontsize=12)

# Show the plot for diabetes=0
plt.show()
```

# Age-Blood Glucose Level

```python
# Age ranges
age_ranges = ['[0, 20)', '[20, 40)', '[40, 60)', '[60, 80)', '[80, 100)']

data_path = "figure_data/age_blood_glucose_level_k250_del50.csv"
df = pd.read_csv(data_path, sep=";")
# Data for each age range
original = df['Original Dataset']
k_anonymity = df['Only K-Anonymity']
random_deletion_average = df['Random Deletion (average)']

# Set the width of the bars
bar_width = 0.2

# Create an array of age range positions for the x-axis
age_range_positions = range(len(age_ranges))

plt.figure(figsize=(8, 6))
# Create the grouped bar chart
original_bars = plt.bar(age_range_positions, original, bar_width, label='Original')
anonymized_bars = plt.bar([x + bar_width for x in age_range_positions], k_anonymity, bar_width, label='Anonymized')
sampling50_bars = plt.bar([x + 2 * bar_width for x in age_range_positions], random_deletion_average, bar_width, label='Sampling50')

# Calculate the centers of the bars
original_centers = [rect.get_x() + rect.get_width() / 2 for rect in original_bars]
anonymized_centers = [rect.get_x() + rect.get_width() / 2 for rect in anonymized_bars]
sampling50_centers = [rect.get_x() + rect.get_width() / 2 for rect in sampling50_bars]

# Draw lines connecting the centers of bars
plt.plot(original_centers, original, marker='o', color='blue')
plt.plot(anonymized_centers, k_anonymity, marker='o', color='green')
plt.plot(sampling50_centers, random_deletion_average, marker='o', color='red')

# Set x-axis labels
plt.xticks([x + bar_width for x in age_range_positions], age_ranges, fontsize=12)
plt.yticks(fontsize=12)

# Set the chart labels and title
plt.xlabel('Age Ranges',fontsize=12)
plt.ylabel('Average Blood Glucose Level',fontsize=12)

# Add a legend
plt.legend(loc='lower right',fontsize=12)

# Rotate x-axis labels for better readability
#plt.xticks(rotation=45)

plt.ylim(0, 180)
# Show the plot
#plt.grid(True)
# Show the plot
plt.show()
```

# Gender - Hypertension

```python
data_path = "figure_data/gender_hypertension_k250_del50.csv"
df = pd.read_csv(data_path, sep=";")
df0 = df.loc[df['hypertension'] == 0]
df1 = df.loc[df['hypertension'] == 1]
# Gender categories
genders = ['Female', 'Male']

# Data for each gender (hypertension=1)
hypertension_1_original = [df1.loc[df1['gender'] == 'Female']['Original Dataset'].iloc[0], 
                           df1.loc[df1['gender'] == 'Male']['Original Dataset'].iloc[0]]
hypertension_1_k_anonymity = [df1.loc[df1['gender'] == 'Female']['Only K-Anonymity'].iloc[0], 
                              df1.loc[df1['gender'] == 'Male']['Only K-Anonymity'].iloc[0]]
hypertension_1_random_deletion_average = [df1.loc[df1['gender'] == 'Female']['Random Deletion (average)'].iloc[0], 
                                          df1.loc[df1['gender'] == 'Male']['Random Deletion (average)'].iloc[0]]

# Data for each gender (hypertension=0)
hypertension_0_original = [df0.loc[df0['gender'] == 'Female']['Original Dataset'].iloc[0], 
                           df0.loc[df0['gender'] == 'Male']['Original Dataset'].iloc[0]]
hypertension_0_k_anonymity = [df0.loc[df0['gender'] == 'Female']['Only K-Anonymity'].iloc[0], 
                              df0.loc[df0['gender'] == 'Male']['Only K-Anonymity'].iloc[0]]
hypertension_0_random_deletion_average = [df0.loc[df0['gender'] == 'Female']['Random Deletion (average)'].iloc[0], 
                                          df0.loc[df0['gender'] == 'Male']['Random Deletion (average)'].iloc[0]]

# Set the width of the bars
bar_width = 0.2

# Create an array of gender positions for the x-axis
gender_positions = np.arange(len(genders))

# Create the grouped bar chart for hypertension=1
plt.figure(figsize=(10, 6))
hypertension_1_original_bars = plt.bar(gender_positions - bar_width, hypertension_1_original, bar_width, label='Original')
hypertension_1_k_anonymity_bars = plt.bar(gender_positions, hypertension_1_k_anonymity, bar_width, label='Anonymized')
hypertension_1_random_deletion_average_bars = plt.bar(gender_positions + bar_width, hypertension_1_random_deletion_average, bar_width, label='Sampling50')

# Calculate the centers of the bars for hypertension=1
hypertension_1_original_centers = [rect.get_x() + rect.get_width() / 2 for rect in hypertension_1_original_bars]
hypertension_1_k_anonymity_centers = [rect.get_x() + rect.get_width() / 2 for rect in hypertension_1_k_anonymity_bars]
hypertension_1_random_deletion_average_centers = [rect.get_x() + rect.get_width() / 2 for rect in hypertension_1_random_deletion_average_bars]

# Draw lines connecting the centers of bars for hypertension=1
plt.plot(hypertension_1_original_centers, hypertension_1_original, marker='o', color='blue')
plt.plot(hypertension_1_k_anonymity_centers, hypertension_1_k_anonymity, marker='o', color='orange')
plt.plot(hypertension_1_random_deletion_average_centers, hypertension_1_random_deletion_average, marker='o', color='green')

# Set x-axis labels
plt.xticks(gender_positions, genders, fontsize=12)

# Set the chart labels and title for hypertension=1
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of people with hypertension', fontsize=12)

# Add a legend for hypertension=1
plt.legend(fontsize=12)

# Show the plot for hypertension=1
plt.show()

# Create the grouped bar chart for hypertension=0
plt.figure(figsize=(10, 6))
hypertension_0_original_bars = plt.bar(gender_positions - bar_width, hypertension_0_original, bar_width, label='Original')
hypertension_0_k_anonymity_bars = plt.bar(gender_positions, hypertension_0_k_anonymity, bar_width, label='Anonymized')
hypertension_0_random_deletion_average_bars = plt.bar(gender_positions + bar_width, hypertension_0_random_deletion_average, bar_width, label='Sampling50')

# Calculate the centers of the bars for hypertension=0
hypertension_0_original_centers = [rect.get_x() + rect.get_width() / 2 for rect in hypertension_0_original_bars]
hypertension_0_k_anonymity_centers = [rect.get_x() + rect.get_width() / 2 for rect in hypertension_0_k_anonymity_bars]
hypertension_0_random_deletion_average_centers = [rect.get_x() + rect.get_width() / 2 for rect in hypertension_0_random_deletion_average_bars]

# Draw lines connecting the centers of bars for hypertension=0
plt.plot(hypertension_0_original_centers, hypertension_0_original, marker='o', color='blue')
plt.plot(hypertension_0_k_anonymity_centers, hypertension_0_k_anonymity, marker='o', color='orange')
plt.plot(hypertension_0_random_deletion_average_centers, hypertension_0_random_deletion_average, marker='o', color='green')

# Set x-axis labels
plt.xticks(gender_positions, genders, fontsize=12)

# Set the chart labels and title for hypertension=0
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of people with no hypertension', fontsize=12)


# Add a legend for hypertension=0
plt.legend(fontsize=12)

# Show the plot for hypertension=0
plt.show()
```
