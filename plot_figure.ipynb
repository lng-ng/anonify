```python
import os
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# Comparing NCP value

```python
data_path = os.path.join('figure_data', 'ncp_table.csv')
df = pd.read_csv(data_path, sep=',')
df.replace(['arx', 'mondrian', 'oka'], ['ARX', 'Mondrian', 'OKA'], inplace=True)
```

```python
sns.set_theme()
plot = sns.lineplot(data=df, x='k', y='ncp', hue='Algorithm')
plot.set(xlabel ="k", ylabel = "NCP(%)")
```
